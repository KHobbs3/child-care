```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities 
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import time
import random
from bs4 import BeautifulSoup
import re
import math
```

```python
opts = Options()
opts.add_argument("user-agent=kt v 1.0")
driver = webdriver.Chrome(ChromeDriverManager().install(), options=opts)
```

```python
my_url = "https://web36.gov.mb.ca/school/school?action=district"
driver.get(my_url)
```

```python
def get_info(eles, i):
    """
    Function to grab information in each back on current page and partition 
    into lists.
    
    """
    eles = driver.find_elements_by_xpath('//*[@id="school"]/tbody/tr[2]/td/div[{}]/descendant::*'.format(i))
    addr.append(eles[1].text.split('\n')[0])
    city.append(eles[1].text.split('\n')[1].split(',')[0])
    pc.append(eles[1].text.split('\n')[2])
    phone.append(eles[1].text.split('\n')[3])
    fax.append(eles[1].text.split('\n')[4])
    grades.append(eles[1].text.split('\n')[5])
    program.append(eles[1].text.split('\n')[6])
#     print(eles[1].text.split('\n')[0])
```

```python
addr = []
city = []
pc = []
phone = []
fax = []
grades = []
program = []
names = []

for option in range(2,52):
    # SELECT DROP DOWN MENU
    driver.find_element_by_xpath('//*[@id="search"]/tbody/tr[1]/td[3]/form/select/option[{}]'.format(option)).click()
    driver.find_element_by_xpath('//*[@id="search"]/tbody/tr[1]/td[3]/form/input').click()
    
    # SWITCH DRIVER TO NEW WINDOW AFTER SELECTING DROP DOWN
    window_after = driver.window_handles[1]
    driver.switch_to.window(window_after)
    
    print("Selecting option {}".format(option-1))
    
    # DETERMINE NUMBER OF TIMES TO CLICK NEXT
    nschool = driver.find_element_by_class_name('n_schools').text
        # number of schools
    m = re.search('[a-z]{2}\s[0-9]{1,3}', nschool)
    num = re.search('[0-9]{1,3}', m.group(0))
    it = int(num.group(0))/10
    nxt = math.ceil(it)
    

    # DETERMINE NUMBER OF FACILITIES ON FIRST PAGE
    # if there are less then 10 then only iterate through the neccessary number of boxes
    # otherwise, grab all ten of them and click next page
    total = driver.find_elements_by_class_name('sc_address')
    toIter = len(total)-1 # not the empty first box
    
    if toIter < 10:
        # 1 -- get names
        sc_names = driver.find_elements_by_class_name('sc_name')
        for n in sc_names[1::]:
            names.append(n.text)
                         
        # 2 -- get info
        for i in range(4,4+toIter):
            get_info(eles, i)

    else:
        # SCRAPE ALL INFO ON PAGE
            # 1 -- get names
        sc_names = driver.find_elements_by_class_name('sc_name')
        for n in sc_names[1::]:
            names.append(n.text)

            # 2 -- get info
        for i in range(4,14):
            get_info(eles, i)

        # SLEEP
        time.sleep(random.uniform(2,8))
        print("Page 1 of {}".format(i, nxt))

        # CLICK NEXT
        for i in range(1,nxt):
            driver.find_element_by_xpath('//*[@id="school"]/tbody/tr[2]/td/div[2]/div/form[2]/input').click()
            print("Page {} of {}".format(i+1, nxt))

            # SWITCH DRIVER TO NEW WINDOW AFTER CLICKING NEXT
            window_after = driver.window_handles[1]
            driver.switch_to.window(window_after)

            # SCRAPE INFO
                # 1 -- get names
            sc_names = driver.find_elements_by_class_name('sc_name')
            for n in sc_names[1::]:
                names.append(n.text)

                # 2 -- get info
                # DETERMINE NUMBER OF FACILITIES ON LAST PAGE
            total = driver.find_elements_by_class_name('sc_address')
            toIter = len(total)-1 # not the empty first box
            for i in range(4,4+toIter):
                get_info(eles, i)

        # SLEEP
        time.sleep(random.uniform(2,8))

driver.close()
```

```python
print(len(names), len(addr), len(city), len(pc), len(phone))
```

```python
mb = pd.DataFrame({'name': names, "address": addr, "city": city, "postal_code": 
              pc, "phone":phone, "fax":fax, "grades":grades, "program":program})
```

```python
mb.drop_duplicates(inplace = True)
```

```python
mb.info()
mb.head()
```

```python
mb.name = mb.name.str.encode('ascii', 'ignore').str.decode('ascii')
```

```python
mb.fax = mb.fax.replace({"Fax: ": ""}, regex = True)
mb.phone = mb.phone.replace({"Phone: ": ""}, regex = True)
mb.grades = mb.grades.replace({"Grades: ": ""}, regex = True)
mb.program = mb.program.replace({"Program: ": ""}, regex = True)
```

```python
mb.to_csv('data/education/MB-education.csv', encoding = "utf-8")
```
